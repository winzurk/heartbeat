```python
import librosa
import librosa.display
import numpy as np
```

```python
y, sr = librosa.load('a0001.wav', sr=2000, mono=True)
```

```python
y = y[0:8000]
```

```python
S = librosa.core.stft(y, n_fft=512, hop_length=20, win_length=100, window='hann', center=True, pad_mode='reflect')
y = librosa.feature.melspectrogram(S=S, sr=2000)
y = librosa.amplitude_to_db(y)
#y = librosa.amplitude_to_db(librosa.feature.mfcc(y=y, sr=sr, n_mfcc=26, n_fft=512, hop_length=128))
#y = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=20)
```

```python
import matplotlib.pyplot as plt
librosa.display.specshow(y, y_axis='log', x_axis='time', sr=2000)
plt.title('Power spectrogram')
plt.colorbar(format='%+2.0f dB')
plt.tight_layout()
plt.show()
```
